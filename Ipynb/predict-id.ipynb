## 1. price agency list

<img src="..\Img\image.png">

- PredictMinCostMonth = digunakan untuk price agency bulanan

<br>

- PredictQuotaCampaigns = digunakan untuk quota agency memperbolehkan buat campaign sebanyak maximal
- PredictMaxCampaigns = digunakan untuk limit maximal agency boleh buat campaign

<br>

- PredictOverageEnable = digunakan untuk enable overage atau tidak
- PredictOverageCampaignsUsed = digunakan untuk mengetahui agency udah buat campaign overage berapa banyak
- PredictOverageRate = digunakan untuk price agency untuk 1 campaign overage

<b>cara charge nya bagaimana?</b>

cara charge nya sebenarnya sangat mudah. harga agency = PredictMinCostMonth + (PredictOverageCampaignsUsed * PredictOverageRate)

<b>gimana skenario overage itu bekerja?</b>

- PredictMinCostMonth = $10
- PredictQuotaCampaigns = 0
- PredictMaxCampaigns = 2
- PredictOverageEnable = true
- PredictOverageCampaignsUsed = 0
- PredictOverageRate = $2

misal gini agency di tanggal 01-Jan-2026 itu sudah agree maka dia di charge sebesar $10. nah di tengah jalan misal di tanggal 15-Jan-2026 dia buat campaign lagi. bagaimana perhitungannya?

charge campaign overage = 
- selisih dari 15-Jan-2026 sampai 31-Jan-2026 = 17 hari
- harga 1 hari untuk overage = $2 / 31 hari = $0.06
- jadi harga untuk 17 hari adalah = $0.06 * 17 hari = $1.02
